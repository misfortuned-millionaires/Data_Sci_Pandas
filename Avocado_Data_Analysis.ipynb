```python
import pandas as pd
import seaborn as sns
```

```python
avo_df = pd.read_csv('avocado.csv')
avo_df.head()
```

```python
avo_df.info()
```

```python
avo_df['Date']=pd.to_datetime(avo_df['Date'])
```

```python
avo_df.info()
```

```python
filt_data = avo_df.query('year == 2015 and type == "conventional" and region == "Albany"')
filt_data.head()
```

```python
avo_melt = pd.melt(avo_df, id_vars='Date',
                   value_vars=['Total Bags', 'Small Bags'],
                   value_name='Count',
                   var_name='Bags')
avo_melt
```

```python
sns.relplot(data=avo_melt, kind='line', x= 'Date', y='Count',
            hue= 'Bags')
```

```python
avo_df['Sales Volume'] = pd.qcut(avo_df['Total Volume'], q=4,
        labels=['Poor','Modest','Good','Excellent'])

avo_df.head()
```

```python
sns.catplot(data=avo_df, kind='count', x='year', hue='Sales Volume')
```

```python
avo_melt2 = pd.melt(avo_df, id_vars='year',
                    value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'],
                    value_name='Count',
                    var_name='Bags')
avo_melt2
```

```python
sns.relplot(data=avo_melt2, kind='line', x='year', y='Count', hue='Bags')
```
